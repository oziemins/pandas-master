```python
import pandas as pd

poland = {
    "cities": ["lodz", "wawa", "krak"],
    "temp": [20, 24, 25],
    "rain": [4, 3, 1]
}

poland_df = pd.DataFrame(poland)
poland_df
```

```python
us = {
    "cities": ["nyc", "la", "washington"],
    "temp": [22, 12, 11],
    "rain": [7, 0, 2]
}

us_df = pd.DataFrame(us)
us_df
```

```python
countries = pd.concat([poland_df, us_df], ignore_index = True)
```

```python
countries
```

```python
countries = pd.concat([poland_df, us_df], keys=["poland", "usa"])
countries
```

```python
countries = pd.concat([poland_df, us_df], axis=1)
countries
```

```python
temp = {
    "city": ["ny", "lodz"],
    "temp": [33, -2]
}
temp_df = pd.DataFrame(temp)
temp_df
```

```python
wind = {
    "city": ["ny", "lodz"],
    "wind": [2, 112]
}
wind_df = pd.DataFrame(wind)
wind_df
```

```python
#concat per 0 rows axis
multi = pd.concat([wind_df, temp_df])
```

```python
multi
```

```python
multi_c = pd.concat([wind_df, temp_df], axis = 1)
multi_c
```

```python
serie_event = pd.Series(["cold", "windy"], name="Event")
df_s = pd.concat([multi_c, serie_event], axis=1)
df_s
```
